eisum:

basically:  
if says `ij,rist->ist`,   
means $C_{ist} = \sum_j \sum_r A_{ij} B_{rist}$  
As long as the subscript not in the right of `->`, do a $\sum$ with that subscript. Order of summation doesn't matter.


```python
"""
eisum syntax and eg
"""
import torch 

# transpose
torch.einsum('ij->ji', [a])
# matrix multiplication
torch.einsum('ik,kj->ij', [a, b])
torch.einsum('bik,bkj->bij', [a, b])    # with batch
torch.einsum('bik,kj->bij', [a, b])     # 2nd matrix broadcast at batch dimension
```
